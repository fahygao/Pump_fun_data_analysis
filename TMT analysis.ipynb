```python
df_before_dilution_per_user_final = df_by_dilution_removed.groupby('OWNER', as_index= False).last().sort_values(by='BLOCK_TIMESTAMP')


df_before_dilution_per_user_final = df_before_dilution_per_user_final[~(df_before_dilution_per_user_final['BALANCE']<=1)].reset_index(drop = True)



```

```python
df_before_dilution_per_user_final
```

```python
# remove the early sniper

df_before_dilution_per_user_final= df_before_dilution_per_user_final[~(df_before_dilution_per_user_final['OWNER'].isin(sniper_wallets))]





```

```python
df_before_dilution_per_user_final = df_before_dilution_per_user_final[['OWNER', 'BALANCE']].iloc[:-1].sort_values(by = 'BALANCE', ascending = False).reset_index(drop = True)
```

```python
df_before_dilution_per_user_final.rename(columns={"OWNER": "WALLET_ADDRESS"})
```

```python
price_token_before_dilution = 0.001975
price_after_dilution = 0.00003451


df_before_dilution_per_user_final['WORTH_IN_USD_BF_DL'] = df_before_dilution_per_user_final['BALANCE'] * price_token_before_dilution




df_before_dilution_per_user_final['WORTH_IN_USD_AF_DL'] = df_before_dilution_per_user_final['BALANCE'] * price_after_dilution




```

```python
df_before_dilution_per_user_final
```

```python
df_before_dilution_per_user_final['LOSS_in_USD'] = df_before_dilution_per_user_final['WORTH_IN_USD_BF_DL'] - df_before_dilution_per_user_final['WORTH_IN_USD_AF_DL'] 


```

```python
initial_amount = 0.03 * 1_000_000_000

df_before_dilution_per_user_final['OWNING_PERCENT'] = df_before_dilution_per_user_final['BALANCE']/initial_amount * 100
```

```python
df_before_dilution_per_user_final
```

```python
total_purchase = df_before_dilution_per_user_final['BALANCE'].sum()
```

```python
total_purchase
```

```python
SOL_price = 135.31

df_before_dilution_per_user_final['LOSS_IN_SOL'] = df_before_dilution_per_user_final['LOSS_in_USD']/SOL_price
```

```python
df_before_dilution_per_user_final=df_before_dilution_per_user_final[['OWNER', 'BALANCE',
                                                                     'OWNING_PERCENT', 'WORTH_IN_USD_BF_DL',
                                                                     'WORTH_IN_USD_AF_DL','LOSS_in_USD', 'LOSS_IN_SOL']]
```

```python
df_before_dilution_per_user_final = df_before_dilution_per_user_final.rename(columns={'OWNER':"WALLET_ADDRESS"})
```
